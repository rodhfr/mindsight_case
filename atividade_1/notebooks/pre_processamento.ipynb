## 2. Diagnóstico de Qualidade

### 2.1 Valores nulos — Tablib Dataset

```python
nulos = df_main.isnull().sum()
pct   = (nulos / len(df_main) * 100).round(1)
resumo_nulos = pd.DataFrame({'nulos': nulos, '% nulos': pct})
resumo_nulos[resumo_nulos['nulos'] > 0].sort_values('% nulos', ascending=False)
```

### 2.2 Valores nulos — performance e área

```python
print('=== performance ===')
print(df_perf.isnull().sum())
print('\n=== área ===')
print(df_area.isnull().sum())
```

### 2.3 Duplicatas

```python
print('Duplicatas Tablib Dataset (linhas completas):', df_main.duplicated().sum())
print('CPFs duplicados em Tablib Dataset           :', df_main['CPF'].duplicated().sum())
print('E-mails duplicados em Tablib Dataset        :', df_main['E-mail'].duplicated().sum())
print('CPFs duplicados em performance              :', df_perf['CPF'].duplicated().sum())
print('CPFs duplicados em área                     :', df_area['CPF'].duplicated().sum())
```

### 2.4 Sobreposição de CPFs entre as abas

Precisamos saber quantos cpf tem em cada aba para entender que tipo de join utilizar.

```python
cpf_main = set(df_main['CPF'])
cpf_perf = set(df_perf['CPF'])
cpf_area = set(df_area['CPF'])

todas    = cpf_main & cpf_perf & cpf_area
so_main  = cpf_main - cpf_perf - cpf_area
so_perf  = cpf_perf - cpf_main
so_area  = cpf_area - cpf_main
main_sem_perf = cpf_main - cpf_perf
main_sem_area = cpf_main - cpf_area

print(f"CPFs no Tablib Dataset  : {len(cpf_main):,}")
print(f"CPFs em performance     : {len(cpf_perf):,}")
print(f"CPFs em área            : {len(cpf_area):,}")
print()
print(f"Nas três abas           : {len(todas):,}")
print(f"Tablib sem performance  : {len(main_sem_perf):,}  → ficam com NaN nos semestres")
print(f"Tablib sem área         : {len(main_sem_area):,}  → ficam sem área mapeada")
print(f"Performance sem Tablib  : {len(so_perf):,}  → perdidos no left join")
print(f"Área sem Tablib         : {len(so_area):,}  → perdidos no left join")
```

### 2.5 Valores únicos de colunas categóricas

```python
print('Cultura classificação:', sorted(df_main['Cultura classificação'].dropna().unique()))
print('Área                 :', sorted(df_area['Área'].unique()))
```
