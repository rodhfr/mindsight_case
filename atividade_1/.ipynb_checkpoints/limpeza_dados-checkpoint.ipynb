# Limpeza de Dados — Atividade 1

Este notebook realiza a limpeza básica do arquivo `Dados - Atividade 1.xlsx`, que contém três abas:
- **Tablib Dataset** — dados de avaliação dos candidatos (4.281 linhas, 58 colunas)
- **performance** — notas de performance por semestre (2.382 linhas)
- **área** — área de cada colaborador (4.394 linhas)

Etapas:
1. Carregamento e inspeção inicial
2. Identificação de problemas
3. Limpeza de cada aba
4. Junção das abas
5. Exportação do dataset limpo

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

FILE = 'Dados - Atividade 1.xlsx'
```

---
## 1. Carregamento e Inspeção Inicial

```python
df_main = pd.read_excel(FILE, sheet_name='Tablib Dataset')
df_perf = pd.read_excel(FILE, sheet_name='performance')
df_area = pd.read_excel(FILE, sheet_name='área')

print(f'Tablib Dataset : {df_main.shape[0]:,} linhas x {df_main.shape[1]} colunas')
print(f'performance    : {df_perf.shape[0]:,} linhas x {df_perf.shape[1]} colunas')
print(f'área           : {df_area.shape[0]:,} linhas x {df_area.shape[1]} colunas')
```

```python
df_main.head(3)
```

```python
df_main.dtypes
```

---
## 2. Diagnóstico de Qualidade

### 2.1 Valores nulos — Tablib Dataset

```python
nulos = df_main.isnull().sum()
pct   = (nulos / len(df_main) * 100).round(1)
resumo_nulos = pd.DataFrame({'nulos': nulos, '% nulos': pct})
resumo_nulos[resumo_nulos['nulos'] > 0].sort_values('% nulos', ascending=False)
```

### 2.2 Valores nulos — performance e área

```python
print('=== performance ===')
print(df_perf.isnull().sum())
print('\n=== área ===')
print(df_area.isnull().sum())
```

### 2.3 Duplicatas

```python
print('Duplicatas Tablib Dataset (linhas completas):', df_main.duplicated().sum())
print('CPFs duplicados em Tablib Dataset           :', df_main['CPF'].duplicated().sum())
print('E-mails duplicados em Tablib Dataset        :', df_main['E-mail'].duplicated().sum())
print('CPFs duplicados em performance              :', df_perf['CPF'].duplicated().sum())
print('CPFs duplicados em área                     :', df_area['CPF'].duplicated().sum())
```

### 2.4 Valores únicos de colunas categóricas

```python
print('Cultura classificação:', sorted(df_main['Cultura classificação'].dropna().unique()))
print('Área                 :', sorted(df_area['Área'].unique()))
```

---
## 3. Limpeza — Tablib Dataset

### 3.1 Remover coluna sem dados úteis

```python
# 'Match' está 100% nula — sem valor analítico
print('Match — % nula:', df_main['Match'].isnull().mean() * 100, '%')
df_main = df_main.drop(columns=['Match'])
```

### 3.2 Remover colunas de URL (não analíticas)

```python
url_cols = [c for c in df_main.columns if c.startswith('URL')]
print('Colunas URL removidas:', url_cols)
df_main = df_main.drop(columns=url_cols)
```

### 3.3 Converter colunas de data/hora para datetime

```python
date_cols = [c for c in df_main.columns if c.startswith('Início') or c.startswith('Fim')]

for col in date_cols:
    df_main[col] = pd.to_datetime(df_main[col], utc=True, errors='coerce')

print('Tipos das colunas de data:')
print(df_main[date_cols].dtypes)
```

### 3.4 Tratar valores ausentes

As colunas de scores (`Potencial Bruto`, `Social`, `Motivacional`, `atributo-*`, `perfil-*`) possuem alta taxa de nulos pois representam testes que nem todos os candidatos realizaram. Esses nulos são **estruturais** — não há como imputá-los sem distorcer a análise. Registramos sua presença e os mantemos como `NaN`.

Para `Cultura pontuação` / `Cultura classificação` (~22,7% nulos) e `Raciocínio` (~24,5% nulos), aplica-se a mesma lógica.

```python
# Criar flag binária indicando quais testes cada candidato realizou
testes = {
    'fez_raciocinio'   : 'Raciocínio',
    'fez_cultura'      : 'Cultura pontuação',
    'fez_social'       : 'Social',
    'fez_motivacional' : 'Motivacional',
    'fez_perfil'       : 'perfil-Comunicação',
    'fez_atributos'    : 'atributo-Comunicação',
}

for flag, col in testes.items():
    df_main[flag] = df_main[col].notna().astype(int)

print('Candidatos que realizaram cada teste:')
df_main[[*testes.keys()]].sum().to_frame('total').T
```

### 3.5 Padronizar coluna de texto: `Cultura classificação`

```python
df_main['Cultura classificação'] = df_main['Cultura classificação'].str.strip().str.title()
print(df_main['Cultura classificação'].value_counts(dropna=False))
```

### 3.6 Padronizar nome: concatenar Nome + Sobrenome

```python
df_main.insert(0, 'Nome Completo',
               (df_main['Nome'].str.strip() + ' ' + df_main['Sobrenome'].str.strip()).str.title())
df_main = df_main.drop(columns=['Nome', 'Sobrenome'])
df_main[['Nome Completo', 'E-mail', 'CPF']].head(3)
```

---
## 4. Limpeza — performance

### 4.1 Renomear colunas para padronizar

```python
df_perf = df_perf.rename(columns={
    'Performance 1º/2019': 'perf_2019_1',
    'Performance 2º/2018': 'perf_2018_2',
    'Performance 1º/2018': 'perf_2018_1',
    'Performance 2º 2017': 'perf_2017_2',
})
df_perf.head(3)
```

### 4.2 Verificar intervalo esperado dos scores de performance

```python
perf_cols = ['perf_2019_1', 'perf_2018_2', 'perf_2018_1', 'perf_2017_2']
df_perf[perf_cols].describe().round(2)
```

---
## 5. Limpeza — área

### 5.1 Padronizar categoria `Área`

```python
df_area['Área'] = df_area['Área'].str.strip().str.title()
print(df_area['Área'].value_counts())
```

### 5.2 Remover duplicatas de CPF em `área`

```python
dup_area = df_area[df_area['CPF'].duplicated(keep=False)]
print(f'Linhas com CPF duplicado em área: {len(dup_area)}')

# Manter a última ocorrência (mais recente)
df_area = df_area.drop_duplicates(subset='CPF', keep='last').reset_index(drop=True)
print(f'Linhas após remoção de duplicatas: {len(df_area)}')
```

---
## 6. Junção das Três Abas

Chave de junção: **CPF**

- `df_main` é o dataset principal (left)
- `df_perf` e `df_area` são enriquecimentos opcionais (left join)

```python
df = (
    df_main
    .merge(df_area, on='CPF', how='left')
    .merge(df_perf,  on='CPF', how='left')
)

print(f'Dataset final: {df.shape[0]:,} linhas x {df.shape[1]} colunas')

# Cobertura das junções
cobertura_area = df['Área'].notna().sum()
cobertura_perf = df['perf_2019_1'].notna().sum()
print(f'Candidatos com área mapeada    : {cobertura_area:,} ({cobertura_area/len(df)*100:.1f}%)')
print(f'Candidatos com performance 2019: {cobertura_perf:,} ({cobertura_perf/len(df)*100:.1f}%)')
```

---
## 7. Resumo Final da Qualidade

```python
nulos_final = df.isnull().sum()
pct_final   = (nulos_final / len(df) * 100).round(1)
resumo_final = pd.DataFrame({'nulos': nulos_final, '% nulos': pct_final})
resumo_final[resumo_final['nulos'] > 0].sort_values('% nulos', ascending=False)
```

```python
print('Tipos de dados do dataset limpo:')
df.dtypes
```

---
## 8. Exportação

```python
OUTPUT = 'dados_limpos.xlsx'

# Excel não suporta datetime com fuso horário — converter para UTC sem timezone
date_cols = [c for c in df.columns if pd.api.types.is_datetime64_any_dtype(df[c])]
for col in date_cols:
    df[col] = df[col].dt.tz_convert('UTC').dt.tz_localize(None)

df.to_excel(OUTPUT, index=False)
print(f'Dataset exportado para "{OUTPUT}" — {df.shape[0]:,} linhas x {df.shape[1]} colunas')
```
